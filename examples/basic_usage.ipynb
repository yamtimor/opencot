# opencot — basic usage

Fetch 5 years of Gold COT data with net positioning and COT index pre-computed.

```python
import opencot as cot

df = cot.get("Gold", report="disaggregated", years=5)
df.tail()
```

```python
# Derived metrics are pre-computed
df[["date", "net_managed_money", "cot_index"]].tail(10)
```

```python
# List all available instruments for a report type
cot.instruments(report="disaggregated")
```
